```python
# Display clean table with "-" for missing values

metric_cols = ["Precision", "Recall", "F1", "AUPRC"]

df_display = df_stage2.copy()

for col in metric_cols:
    df_display[col] = df_display[col].apply(
        lambda x: "-" if pd.isna(x) else f"{x:.2f}"
    )

df_display = df_display.sort_values(
    ["Dataset", "F1"],
    ascending=[True, False],
    na_position="last"
).reset_index(drop=True)

df_display
```

```python
# ==========================================================
# Highlight best and second-best values per dataset and metric
# Best = bold, Second-best = underline
# Missing values = "-"
# ==========================================================

metric_cols = ["Precision", "Recall", "F1", "AUPRC"]

def format_metric_table(group):
    group = group.copy()

    for metric in metric_cols:
        values = group[metric].dropna().sort_values(ascending=False).unique()

        best = values[0] if len(values) >= 1 else None
        second = values[1] if len(values) >= 2 else None

        formatted_values = []

        for x in group[metric]:
            if pd.isna(x):
                formatted_values.append("-")
            elif best is not None and np.isclose(x, best):
                formatted_values.append(r"\textbf{" + f"{x:.2f}" + "}")
            elif second is not None and np.isclose(x, second):
                formatted_values.append(r"\underline{" + f"{x:.2f}" + "}")
            else:
                formatted_values.append(f"{x:.2f}")

        group[metric] = formatted_values

    return group

df_latex_highlighted = (
    df_stage2
    .groupby("Dataset", group_keys=False)
    .apply(format_metric_table)
)

# Sort by dataset and F1, keeping missing values last
df_latex_highlighted["_F1_sort"] = df_stage2["F1"]

df_latex_highlighted = (
    df_latex_highlighted
    .sort_values(["Dataset", "_F1_sort"], ascending=[True, False], na_position="last")
    .drop(columns=["_F1_sort"])
    .reset_index(drop=True)
)

df_latex_highlighted
```

```python
# ==========================================================
# Journal-style grouped LaTeX table
# Rows: Models
# Column groups: Datasets
# Subcolumns: P, R, F1, AUPRC
# Best = bold, Second-best = underline
# Missing = "-"
# ==========================================================

import pandas as pd
import numpy as np

# ----------------------------------------------------------
# Stage 2 results
# Values are percentages.
# Missing values are np.nan and will be displayed as "-".
# MTAD-Kanformer was removed because AUPRC was not available
# under the same evaluation protocol.
# ----------------------------------------------------------

data = [
    # Model, Dataset, Precision, Recall, F1, AUPRC

    # Proposed model
    ["DeepSKAN", "SWaT", 93.52, 96.69, 95.08, 98.10],
    ["DeepSKAN", "MSL",  91.58, 93.27, 92.42, 90.47],
    ["DeepSKAN", "SMAP", 89.24, 69.96, 78.43, 87.36],

    # DeepSVDD alone
    ["DeepSVDD", "SWaT", 26.11, 84.73, 39.92, 24.29],
    ["DeepSVDD", "MSL",  91.80, 65.67, 76.57, 77.61],
    ["DeepSVDD", "SMAP", 13.63, 89.63, 23.65, 15.23],

    # KAN-AD
    ["KAN-AD", "SWaT", 73.05, 89.59, 74.15, 69.33],
    ["KAN-AD", "MSL",  53.10, 66.73, 46.33, 36.67],
    ["KAN-AD", "SMAP", 91.59, 44.91, 57.04, 50.15],

    # DLinear
    ["DLinear", "SWaT", 95.53, 95.77, 95.65, 97.98],
    ["DLinear", "MSL",  90.95, 86.60, 88.72, 91.48],
    ["DLinear", "SMAP", 71.80, 71.10, 71.45, 80.77],

    # LSTM Autoencoder
    ["LSTM-AE", "SWaT", 94.05, 79.33, 86.06, 86.70],
    ["LSTM-AE", "MSL",  29.41, 75.55, 42.34, 31.59],
    ["LSTM-AE", "SMAP", 66.57, 52.92, 58.97, 55.13],

    # KARMAD
    ["KARMAD", "SWaT", 90.68, 77.96, 83.84, 86.65],
    ["KARMAD", "MSL",  85.54, 80.67, 83.03, 83.11],
    ["KARMAD", "SMAP", 94.50, 54.74, 69.32, 69.37],
]

df_stage2_latex = pd.DataFrame(
    data,
    columns=["Model", "Dataset", "Precision", "Recall", "F1", "AUPRC"]
)

# ----------------------------------------------------------
# Table formatting
# ----------------------------------------------------------

metric_map = {
    "Precision": "P",
    "Recall": "R",
    "F1": "F1",
    "AUPRC": "AUPRC"
}

datasets_order = ["SWaT", "MSL", "SMAP"]

models_order = [
    "DeepSVDD",
    "KAN-AD",
    "DLinear",
    "LSTM-AE",
    "KARMAD",
    "DeepSKAN",
]

def format_best_second(df, dataset, metric, model):
    subset = df[df["Dataset"] == dataset][["Model", metric]].dropna()
    row = df[(df["Dataset"] == dataset) & (df["Model"] == model)]

    if row.empty or pd.isna(row.iloc[0][metric]):
        return "-"

    x = row.iloc[0][metric]
    value = f"{x:.2f}"

    values = np.sort(subset[metric].unique())[::-1]
    best = values[0] if len(values) >= 1 else None
    second = values[1] if len(values) >= 2 else None

    if best is not None and np.isclose(x, best):
        return r"\textbf{" + value + "}"
    elif second is not None and np.isclose(x, second):
        return r"\underline{" + value + "}"
    else:
        return value

# ----------------------------------------------------------
# Build grouped table
# ----------------------------------------------------------

rows = []

for model in models_order:
    row = {"Model": model}

    for dataset in datasets_order:
        for metric, short_name in metric_map.items():
            row[(dataset, short_name)] = format_best_second(
                df_stage2_latex,
                dataset,
                metric,
                model
            )

    rows.append(row)

df_grouped = pd.DataFrame(rows)

ordered_cols = ["Model"] + [
    (dataset, short_name)
    for dataset in datasets_order
    for short_name in metric_map.values()
]

df_grouped = df_grouped[ordered_cols]

multi_cols = [("", "Model")] + [
    (dataset, short_name)
    for dataset in datasets_order
    for short_name in metric_map.values()
]

df_grouped.columns = pd.MultiIndex.from_tuples(multi_cols)

df_grouped
```

```python
latex_table = df_grouped.to_latex(
    index=False,
    escape=False,
    multicolumn=True,
    multicolumn_format="c",
    column_format="l" + "c" * (len(datasets_order) * len(metric_map)),
    caption=r"Main benchmark comparison on SWaT, MSL, and SMAP. Best results are shown in bold and second-best results are underlined.",
    label="tab:main_benchmark_comparison"
)

print(latex_table)
```
